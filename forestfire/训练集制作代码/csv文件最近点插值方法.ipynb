```python
###临近插值程序。 作者：dyl_hx 联系：1603115627
##更新于20231024

import os
from datetime import datetime, timedelta
import xarray as xr
import csv
import numpy as np


with open('/public/home/duyl_hx/firedata/fire2010-2017_new.csv', mode='r') as file:
    reader = csv.reader(file)
    header = next(reader) 
    data = list(reader)
    
header.append("可燃物含水率")

###根据文件名获取日期
indata_path = '/public/home/duyl_hx/firedata/a卫星反演/a可燃物含水率/test'

# 读入火点数据
filtered_data = []

for subdir, dirs, files in os.walk(indata_path):
    for file in files:
        if file.endswith('.nc'):
            ##依据文件名提取年和周
            year = int(file[:4])
            week = int(file[file.find('w')+1:file.find('.nc')])
            ##根据提取到的年和周计算对应文件所属日期
            start_date = datetime(year, 1, 1) + timedelta(weeks=week - 1)
            print("开始时间",start_date)
            end_date = datetime(year, 1, 1) + timedelta(weeks=week)
            if end_date.year != year:
                end_date = datetime(year, 12, 31)  # 设置为年的最后一天
            print("结束时间",end_date)
            #读入可燃物数据
            file_path = os.path.join(subdir, file)
            print("正在处理文件",file_path)
            ds = xr.open_dataset(file_path)
            
            for row in data:
                image_date = row[2]  # 日期在第三列
                date = datetime.strptime(image_date, '%Y-%m-%d %H:%M:%S')
                if start_date <= date <= end_date:
                    print("处理日期",date)
                    lon1 = float(row[3])  # 经度在第四列
                    lat1 = float(row[4])  # 纬度在第五列
                    # 读入 xarray 数据集
                    # 选择最接近的经度和纬度
                    ds_subset = ds.sel(lat=lat1, lon=lon1, method='nearest')
                    #print("插值点",ds_subset)
                    # 找到最接近点的索引
                    nearest_lat = ds_subset.lat.values
                    nearest_lon = ds_subset.lon.values
                    # 获取最接近点的值
                    nearest_value = ds.Band1.sel(lat=nearest_lat, lon=nearest_lon, method='nearest').values
                    # 将最接近值添加到原始数据的新列中
                    row.append(nearest_value)
                    # 将原始行添加到筛选数据中
                    filtered_data.append(row)
print(filtered_data)
# 将筛选后的数据保存为新的CSV文件
with open('/public/home/duyl_hx/firedata/20231025.csv', mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(header)  # 写入标题行
    writer.writerows(filtered_data)
    
print("插值完成:")
```

```python
import os
from datetime import datetime, timedelta
import xarray as xr
import csv
import numpy as np

def interpolate_and_save(interp_data_path, data_path, new_column_name, output_file_path):
    # 读取火点数据
    with open(interp_data_path, mode='r') as file:
        reader = csv.reader(file)
        header = next(reader)
        data = list(reader)

    header.append(new_column_name)

    # 读入可燃物数据
    filtered_data = []

    for subdir, dirs, files in os.walk(data_path):
        for file in files:
            if file.endswith('.nc'):
                year = int(file[:4])
                week = int(file[file.find('w')+1:file.find('.nc')])
                start_date = datetime(year, 1, 1) + timedelta(weeks=week - 1)
                print("开始时间", start_date)
                end_date = datetime(year, 1, 1) + timedelta(weeks=week)
                if end_date.year != year:
                    end_date = datetime(year, 12, 31)  # 设置为年的最后一天
                print("结束时间", end_date)

                file_path = os.path.join(subdir, file)
                print("正在处理文件", file_path)
                ds = xr.open_dataset(file_path)

                for row in data:
                    image_date = row[2]  # 日期在第三列
                    date = datetime.strptime(image_date, '%Y-%m-%d %H:%M:%S')
                    if start_date <= date <= end_date:
                        print("处理日期", date)
                        lon1 = float(row[3])  # 经度在第四列
                        lat1 = float(row[4])  # 纬度在第五列
                        # 选择最接近的经度和纬度
                        ds_subset = ds.sel(lat=lat1, lon=lon1, method='nearest')
                        # 找到最接近点的索引
                        nearest_lat = ds_subset.lat.values
                        nearest_lon = ds_subset.lon.values
                        # 获取最接近点的值
                        nearest_value = ds.Band1.sel(lat=nearest_lat, lon=nearest_lon, method='nearest').values
                        # 将最接近值添加到原始数据的新列中
                        row.append(nearest_value)
                        # 将原始行添加到筛选数据中
                        filtered_data.append(row)

    # 将筛选后的数据保存为新的CSV文件
    with open(output_file_path, mode='w', newline='') as file:
        writer = csv.writer(file)
        writer.writerow(header)  # 写入标题行
        writer.writerows(filtered_data)

    print("插值完成")

# 使用示例
interp_data_path = '/public/home/duyl_hx/firedata/fire2010-2017_new.csv'
data_path = '/public/home/duyl_hx/firedata/a卫星反演/a可燃物含水率/test'
new_column_name = "可燃物含水率"  # 新增列名
output_file_path = '/public/home/duyl_hx/firedata/202310251.csv'  # 输出文件路径

interpolate_and_save(interp_data_path, data_path, new_column_name, output_file_path)
```
